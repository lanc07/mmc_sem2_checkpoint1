```python
# Importando a biblioteca
import numpy as np
```

```python
# Criando vetores arbitrários
u = np.array([1,2])
v = np.array([4,5])
w = np.array([7,8])
```

```python
# Criando escalares
a = 7
b = 23
```

```python
# Associativa da adição
print(u + (v + w))
print((u + v) + w)
```

```python
# Comutatividade da adição
print(u + v)
print(v + u)
```

```python
# Elemento identidade da adição
print(v + 0)
print(v)
```

```python
# Elemento inverso da adição
print(v + (-v))
```

```python
# Compatibilidade da multiplicação por escalar com a multiplicação do corpo
print(a * (b * v))
print((a * b) * v)
```

```python
# Elemento identidade da multiplicação por escalar
print(1 * v)
print(v)
```

```python
# Distributividade da multiplicação por escalar em relação è adição de vetores
print(a * (u + v))
print(a * u + a * v)
```

```python
# Distributividade da multiplicação por escalar em relação a adição do corpo
print((a + b) * v)
print(a * v + b * v)
```
